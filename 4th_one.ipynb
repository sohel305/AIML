```python
years=range(2000,2008)

apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('year')
plt.ylabel('yield')
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('year')
plt.ylabel('yield')
plt.title('Crop yield in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')

plt.ylabel('yield')
plt.title('Crop yield in kanto')
plt.legend(['apples','oranges'])
```

```python
plt.plot(years,apples,marker='s', c='b', ls='-', lw=2, ms=8, mec='navy')
plt.plot(years,oranges,marker='x', c='r', ls='--', lw=3, ms=10, alpha=0.5)

plt.xlabel('year')
plt.ylabel('yield')

plt.title('Crop yield in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')

plt.xlabel('year')
plt.ylabel('yield')

plt.title('Crop yield in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(years,oranges,'or')
plt.title('yield of oranges')
```

```python
import matplotlib.pyplot as plt

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]

plt.bar(x,y)

plt.title('bar Chart')

plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]

plt.scatter(x,y)

plt.legend("A")
plt.xlabel("x")
plt.ylabel("y")

plt.title("Scatter chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
e=(0.10,0.01,0.01,0.01)

plt.pie(x, explode =e)

plt.title("Pie chart")
plt.show()
```
